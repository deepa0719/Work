#  Edit Distance

```python
def editDistance(str1,str2):
    l1=len(str1)
    l2=len(str2)
    table=[[0]*(l2+1) for _ in range(l1+1)]
    
    for i in range(l1+1):
        table[i][0]=i
    for j in range(l2+1):
        table[0][j]=j

    for i in range(1,l1+1):
        for j in range(1,l2+1):
            if str1[i-1]==str2[j-1]:
                table[i][j]=table[i-1][j-1]
  
            else:
                table[i][j]=1+min(table[i-1][j],table[i][j-1],table[i-1][j-1])

    
    for i in range(l1+1):
        for j in range(l2+1):
            print(table[i][j],end='\t')
        print("\n")
        
    print("Edit Distance=",table[-1][-1],"\n")
    
    i=l1
    j=l2
    d=table[-1][-1]
    while i>0:
        while j>0:
            if str1[i-1]==str2[j-1]:
                print(str1[i-1]," matches" ,str2[j-1],"=",d)   
                i-=1
                j-=1
            else:
                n=min(table[i-1][j],table[i][j-1],table[i-1][j-1])     
                if n == table[i-1][j-1]:
                    d-=1
                    print("Replacement = 2")
                    print(str1[i-1]," doesn't match", str2[j-1],"=",d)
                    i-=1
                    j-=1
                elif n == table[i][j-1]:
                    d-=1
                    print("Insertion = 1")
                    print(str1[i-1]," doesn't match" , str2[j-1],"=",d)
                    j-=1
                elif n == table[i-1][j]:
                    d-=1
                    print("Deletion = 1")
                    print(str1[i-1]," doesn't match" , str2[j-1],"=",d)
                    i-=1
```

```python
str2=input("Enter String 1 ")
str1=input("Enter String 2 ")
editDistance(str1,str2)
```

# Jaro Distance

```python
def jaro_distance(str1, str2):
    s1_len = len(str1)
    s2_len = len(str2)
 
    if s1_len == 0 and s2_len == 0:
        return 1
 
    match_distance = (max(s1_len, s2_len) // 2) - 1
 
    s_matches = [False] * s1_len
    t_matches = [False] * s2_len
 
    matches = 0
    transpositions = 0
 
    for i in range(s1_len):
        start = max(0, i - match_distance)
        end = min(i + match_distance + 1, s2_len)
 
        for j in range(start, end):
            if t_matches[j]:
                continue
            if str1[i] != str2[j]:
                continue
            s_matches[i] = True
            t_matches[j] = True
            matches += 1
            break
 
    if matches == 0:
        return 0
 
    k = 0
    for i in range(s1_len):
        if not s_matches[i]:
            continue
        while not t_matches[k]:
            k += 1
        if str1[i] != str2[k]:
            transpositions += 1
        k += 1
    jaro_sim=((matches / s1_len) + (matches / s2_len) + ((matches - transpositions / 2) / matches)) / 3
    jaro_dist=1-jaro_sim
    print("Jaro Similarity=",jaro_sim)
    print("Jaro Distance=",jaro_dist)
```

```python
str2=input("Enter String 1 ")
str1=input("Enter String 2 ")
jaro_distance(str1,str2)
```

# Dice N-Gram Distance

```python
def dice_ngram_distance(a,b):
    if not len(a) or not len(b): return 0.0
    if len(a) == 1:  a=a+u'.'
    if len(b) == 1:  b=b+u'.'
    
    a_bigram_list=[]
    for i in range(len(a)-1):
      a_bigram_list.append(a[i:i+2])
    b_bigram_list=[]
    for i in range(len(b)-1):
      b_bigram_list.append(b[i:i+2])
      
    a_bigrams = set(a_bigram_list)
    b_bigrams = set(b_bigram_list)
    overlap = len(a_bigrams & b_bigrams)
    dice_coeff = overlap * 2.0/(len(a_bigrams) + len(b_bigrams))
    return dice_coeff
```

```python
str2=input("Enter String 1 ")
str1=input("Enter String 2 ")
print(dice_ngram_distance(str1,str2))
```

# LCS

```python
def lcs(S1, S2):
    m=len(S1)
    n=len(S2)
    L = [[0 for x in range(n+1)] for x in range(m+1)]

    for i in range(m+1):
        for j in range(n+1):
            if i == 0 or j == 0:
                L[i][j] = 0
            elif S1[i-1] == S2[j-1]:
                L[i][j] = L[i-1][j-1] + 1
            else:
                L[i][j] = max(L[i-1][j], L[i][j-1])

    index = L[m][n]

    lcs_algo = [""] * (index+1)
    lcs_algo[index] = ""

    i = m
    j = n
    while i > 0 and j > 0:

        if S1[i-1] == S2[j-1]:
            lcs_algo[index-1] = S1[i-1]
            i -= 1
            j -= 1
            index -= 1

        elif L[i-1][j] > L[i][j-1]:
            i -= 1
        else:
            j -= 1
            
    print("S1 : " + S1 + "\nS2 : " + S2)
    print("LCS: " + "".join(lcs_algo))
    print("length",len(lcs_algo)-1)
```

```python
X=input("Enter String 1 ")
Y=input("Enter String 2 ")

lcs(X, Y)
```
