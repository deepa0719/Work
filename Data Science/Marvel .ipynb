```python
#loading required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
marvel=pd.read_csv("Datasets/charcters_stats.csv")
```

```python
marvel.head()
```

```python
marvel.shape
```

```python
marvel['Alignment'].value_counts()
```

```python
heroes = marvel[marvel['Alignment']=='good']
```

```python
heroes.describe()
```

```python
heroes[['Name','Speed']].sort_values(by=['Speed'],ascending=False).head()
```

```python
superspeed = heroes[heroes['Speed']==100]
superspeed.shape
```

```python
slowest = heroes[heroes['Speed']==1]
slowest.shape
```

```python
smart_heroes = heroes[['Name','Intelligence','Combat']].sort_values(by=['Intelligence'],ascending=False)
smart_heroes.head()
```

```python
max_power = heroes[heroes['Power']==100].sort_values(by=['Total'],ascending=False)
max_power.head()
```

```python
plt.figure(figsize=(10,6))
plt.bar(list(max_power['Name'])[0:5],list(max_power['Total'])[0:5],color="turquoise")
plt.show()
```

```python
villains = marvel[marvel['Alignment']=='bad']
```

```python
villains.head()
```

```python
villains.describe()
```

```python
villains.sort_values(by=['Intelligence'],ascending=False).head()
```

```python
bad_max = villains.sort_values(by=['Total'],ascending=False)
```

```python
villains[villains['Name']=='Dormammu']
```

```python
plt.figure(figsize=(7,5))
plt.bar(list(bad_max['Name'])[0:5],list(bad_max['Total'])[0:5], color="magenta")
plt.show()
```

```python
#Comparing speeds of superheroes and villains
#Spacing between the two subplots
plt.subplots_adjust(left=None, bottom=None, right=2, top=None, wspace=1, hspace=None)
#plot1
plt.subplot(1, 2, 1)  # 1 line, 2 rows, index nr 1 (first position in the subplot)
plt.hist(heroes['Speed'],edgecolor='black',color='palegreen')
plt.gca().set_title('Heroes speed distribution',fontsize=20)
plt.xlabel('Speed',fontsize=15)
plt.ylabel('Frequency', fontsize=15)


#plot2
plt.subplot(1, 2, 2)  # 1 line, 2 rows, index nr 2 (second position in the subplot)
plt.hist(villains['Speed'],edgecolor='black',color='mediumorchid')
plt.gca().set_title('Villains speed distribution',fontsize=20)
plt.xlabel('Speed',fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.show()
```

```python
#Comparing combats of superheroes and villains
#Spacing between the two subplots
plt.subplots_adjust(left=None, bottom=None, right=2, top=None, wspace=1, hspace=None)
#plot1
plt.subplot(1, 2, 1)  # 1 line, 2 rows, index nr 1 (first position in the subplot)
plt.hist(heroes['Combat'],edgecolor='black',color='palegreen')
plt.gca().set_title('Superheroes combat distribution',fontsize=20)
plt.xlabel('Number of Combats',fontsize=15)
plt.ylabel('Frequency', fontsize=15)


#plot2
plt.subplot(1, 2, 2)  # 1 line, 2 rows, index nr 2 (second position in the subplot)
plt.hist(villains['Combat'],edgecolor='black',color='mediumorchid')
plt.gca().set_title('Villains combat distribution',fontsize=20)
plt.xlabel('Number of Combats',fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.show()
```
